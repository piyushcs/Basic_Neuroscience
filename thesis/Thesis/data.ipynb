### Following function will load the data then normalize it to 0-255 so that we can have a image representation on the data:

```python
data = loadmat('/Users/piyush.kumar/Documents/Archive/fMRI_project/data/data-science-P1.mat')
raw_data = []
cond_number = []

# print data['meta'][0][0][7]
word = data['info']['word'][0]
# print data['info'][0][0]
# print data['meta']

words = list()
for i in word:
    words.append(str(i[0]))

for i in range(360):
    temp = np.array(data['data'][i][0][0])
    raw_data.append(temp)

r = np.array(raw_data)
og_data = np.array(raw_data)

print og_data[0]

np.save('data_p1.npy', og_data)

print og_data[0]

smin = 0
smax = 1

x = np.array(raw_data)

xmax, xmin = x.max(), x.min()

norm_data = (x - xmin)*(smax - smin)/(xmax - xmin) + smin

print norm_data.shape
```

### Next function will convert the image shape from (21764, ) to (51, 61, 23) to generate 3D voxels. For now it is only for first image, there are total 360 images.

```python
np_norm_data = np.asarray(norm_data)
words = np.asarray(words)
np.save('data_p1', norm_data)
np.save('labels_p1', words)
```

```python
norm_3D_data = list()
val_index = loadmat('/Users/piyush.kumar/Documents/Archive/fMRI_project/data/val.mat')
val = val_index['X']

for n in range(360):
    np_val = np.array(val, dtype=float)
    img = np_norm_data[n]
    print '{0}th element is converted to 3D'
    for i in range(val.shape[0]):
        for j in range(val.shape[1]):
            for k in range(val.shape[2]):
                index = val[i][j][k]
                if index != 0:
                    np_val[i][j][k] = img[index-1]
    norm_3D_data.append(np_val)
```

```python
norm_3D_data = np.asarray(norm_3D_data)
words = np.asarray(words)

np.save('data_p1_3D', norm_3D_data)
# np.save('labels', words)
```
