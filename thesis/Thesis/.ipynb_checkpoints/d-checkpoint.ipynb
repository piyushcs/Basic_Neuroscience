```python
import numpy as np
from scipy.io import loadmat
```

### Following function will load the data then normalize it to 0-255 so that we can have a image representation on the data:

```python
data_test = loadmat('/Users/piyush.kumar/Documents/Archive/fMRI_project/data/data-science-P1.mat')

smin = 0
smax = 255

norm_data = list()

for i in range(len(data_test['data'])):
    X = data_test['data'][i][0][0]
    X = (X - min(X))*(smax - smin)/(max(X) - min(X)) + smin
    norm_data.append(X)
```

### Next function will convert the image shape from (21764, ) to (51, 61, 23) to generate 3D voxels. For now it is only for first image, there are total 360 images.

```python
np_norm_data = np.asarray(norm_data)

val = loadmat('/Users/piyush.kumar/Documents/Archive/fMRI_project/data/val.mat')
val = val['X']

print val.shape
img_1 = np_norm_data[1]

for i in range(val.shape[0]):
    for j in range(val.shape[1]):
        for k in range(val.shape[2]):
            index = val[i][j][k]
            if index != 0:
                val[i][j][k] = img_1[index-1]
```

### Now we have first fMRI image in the shape of 3D voxel
